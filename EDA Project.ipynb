```python
#setting up figure size
plt.figure(figsize=(16,40))
sns.set_style("whitegrid")

plt.subplot(6,2,1)
category_wise_sales = df.groupby('Category')['Sales'].sum()
diff_col = ['red','green','blue']
plt.bar(category_wise_sales.index,category_wise_sales.values,color='skyblue')
plt.title('Total Size By Category')
plt.ylabel('Sales')

#pie chart
plt.subplot(6,2,2)
Region_counts=df['Region'].value_counts()
plt.pie(Region_counts.values, labels = Region_counts.index, autopct="%0.1f%%")
plt.title('Total Order By Region')

#Scatter plot
plt.subplot(6,2,3) 
sns.scatterplot(data=df,x='Sales',y='Profit',hue='Category')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

#Box plot
plt.subplot(6,2,4)
sns.boxplot(data=df,x='Segment',y='Profit')
plt.title('Profit Distributed By segment')

#Correlation heat map
plt.subplot(6,2,5)
sns.heatmap(df.corr(numeric_only=True),annot=True,vmin=-1,vmax=1,cmap='coolwarm' )

#scatterplot
plt.subplot(6,2,6)
sns.scatterplot(data=df,x='Discount',y='Profit')
plt.title('Discount vs Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')


#Line plot
plt.subplot(6,2,7)
monthly_sales=df.resample('ME',on='Order Date')['Sales'].sum()
plt.plot(monthly_sales.index,monthly_sales.values)
plt.title('Monthly Sales')
plt.xlabel('Months')
plt.ylabel('Total Sales')


#Hisogram
plt.subplot(6,2,8)
plt.hist(df['Quantity'],bins=10,color='Orange')
plt.title('Profit Distributin By The Category')


#Count plot
plt.subplot(6,2,9)
sns.countplot(df['Sub-Category'],color='pink')
plt.title('Sub_Category Countplot')


#Violin plot
plt.subplot(6,2,10)
sns.violinplot(data=df,x='Region',y='Quantity',color='green')
xlabel= 'Region'
ylabel='Quantity'
plt.title('Most Selling Products')

#kde plot
plt.subplot(6,2,11)
sns.kdeplot(df['Quantity'],color='violet')
plt.title('Number of Quantity')


#displot
plt.subplot(6,2,12)
sns.displot(df['Sales'][df['Sales']<250],color='Red')
plt.title('Below 250 Sales')

plt.tight_layout()
plt.show()
```
